```python
# advanced grammar score test
from grammar_score import AdvancedGrammarScorer
import re

sentences = [sentence for sentence in sentences if not re.search(r"[가-힣]", sentence)]

score_c = AdvancedGrammarScorer()
scores = []
sentences_text = []
test = {}

for s in sentences:
    score, detail = score_c.score_sentence(s)
    scores.append(score)
    sentences_text.append(s)
    test[s] = score
# print(test)
```
