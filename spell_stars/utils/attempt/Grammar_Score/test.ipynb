```python
sentences = [
    # 완벽한 문장 3개
    "The cat is sleeping on the bed.",
    "She finished her homework before dinner.",
    "They went to the beach and enjoyed the sunny weather.",
    # 비문 (단계별로 오류 증가)
    # 단계 1: 가벼운 오류
    "She go to the store yesterday.",
    "She went to the store yesterday.",  # 시제와 주어-동사 일치 오류 ("go"를 "went"로 변경 필요)
    # 단계 2: 중간 정도 오류
    "udfdfsdfsdfsdf don't likes the apples because it soury.",  # 주격 대명사 오류 ("Him" -> "He"), "don't" 대신 "doesn't" 필요, 단수형 "it" 대신 복수형 "they" 필요, "soury"는 잘못된 단어
    # 단계 3: 심각한 오류
    "Yesterday me and him was going store for buying milk.",
    "asdfasdfasdfasdfasdfasdfasdf",
    "I reeeeerreee erererererdflksjdfsadf.",
    "I dont dont dont dont",
    "i dont dont dont dont",
]
```

```python
# advanced grammar score test
from grammar_score import AdvancedGrammarScorer

score_c = AdvancedGrammarScorer()
scores = []
sentences_text = []
test = {}

for s in sentences:
    score, detail = score_c.score_sentence(s)
    scores.append(score)
    sentences_text.append(s)
    test[s] = score
print(test)
```

```python
# 최종 문장 필터

filtered_sentence = {
    sentence: score for sentence, score in test.items() if score < 0.8
}  # 값을 조절, 4분위 수로 해도 됨
filtered_sentence
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(scores, kde=True)
plt.title("Dist")
plt.show()
```
